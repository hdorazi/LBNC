```python
from google.colab import files
from konlpy.tag import Okt

# 파일 업로드
uploaded = files.upload()

# 형태소 분석기 초기화
okt = Okt()

# 가사 데이터 읽기 및 형태소 분석
lyrics = []
for filename in uploaded.keys():
    with open(filename, "r", encoding="utf-8") as f:
        lyrics.append(f.read())

# 형태소 분석
def analyze_lyrics(lyrics):
    results = []
    for idx, lyric in enumerate(lyrics):
        print(f"분석 중: 가사 {idx+1}")
        morphs = okt.morphs(lyric)  # 형태소
        nouns = okt.nouns(lyric)   # 명사
        pos = okt.pos(lyric)       # 품사 태깅
        results.append({
            "가사 번호": idx+1,
            "형태소": morphs,
            "명사": nouns,
            "품사": pos
        })
    return results

# 분석 실행
analysis_results = analyze_lyrics(lyrics)

# 결과 출력
for result in analysis_results:
    print(f"\n가사 {result['가사 번호']} 분석 결과:")
    print("형태소:", result["형태소"])
    print("명사:", result["명사"])
    print("품사 태깅:", result["품사"])
```

```python
import re

# 특수문자 제거 함수
def clean_lyrics(text):
    return re.sub(r"[^\w\s]", "", text)

# 가사 전처리
cleaned_lyrics = [clean_lyrics(lyric) for lyric in lyrics]
print(cleaned_lyrics)
```

```python
from collections import Counter

# 모든 가사를 형태소 단위로 분리
all_morphs = []
for lyric in lyrics:
    all_morphs.extend(okt.morphs(lyric))

# 빈도 계산
counter = Counter(all_morphs)
print("가사 전체 키워드 빈도:", counter.most_common(10))
```

```python
from konlpy.tag import Okt
from itertools import combinations
from collections import Counter

# 형태소 분석기 초기화
okt = Okt()

# 중복 제거 함수
def remove_duplicates(texts):
    return list(set(texts))

# 특정 품사 추출 함수 (명사, 동사, 형용사 포함)
def extract_target_pos(texts, target_pos=("Noun", "Verb", "Adjective")):
    words_list = []
    for text in texts:
        words = [word for word, pos in okt.pos(text) if pos in target_pos]
        words_list.append(words)
    return words_list

# 공기어 추출 함수
def extract_cooccurrences(words_list):
    cooccurrences = []
    for words in words_list:
        cooccurrences.extend(combinations(words, 2))  # 단어쌍 생성
    return cooccurrences

# 중복 제거, 품사 추출, 공기어 분석
def analyze_lyrics(lyrics):
    # 중복 제거
    unique_lyrics = remove_duplicates(lyrics)
    print(f"중복 제거 완료! 총 {len(unique_lyrics)}개의 고유한 가사가 남았습니다.")

    # 형태소 분석 및 공기어 추출
    target_words = extract_target_pos(unique_lyrics, target_pos=("Noun", "Verb", "Adjective"))
    cooccurrences = extract_cooccurrences(target_words)
    cooccurrence_counts = Counter(cooccurrences)

    return cooccurrence_counts

# 공기어 분석 실행
cooccurrence_counts = analyze_lyrics(lyrics)

# 공기어 빈도 상위 10개 출력
print("공기어 분석 결과 (중복 제거 후):")
print(cooccurrence_counts.most_common(10))
```

```python
import pandas as pd

# 공기어 분석 결과를 엑셀로 저장
def save_cooccurrence_to_excel(cooccurrence_counts, output_file="cooccurrence_results.xlsx"):
    # 단어쌍과 빈도를 데이터프레임으로 변환
    data = [{"단어1": word1, "단어2": word2, "빈도": freq}
            for (word1, word2), freq in cooccurrence_counts.items()]
    df = pd.DataFrame(data)

    # 엑셀 파일로 저장
    df.to_excel(output_file, index=False, engine='openpyxl')
    print(f"공기어 분석 결과가 '{output_file}' 파일로 저장되었습니다.")

# 엑셀 저장 실행
save_cooccurrence_to_excel(cooccurrence_counts)
```
